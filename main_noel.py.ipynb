```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

url="https://raw.githubusercontent.com/johnosullivan/ml-horse-prediction/master/data/training-2016-10-01-2016-12-31.csv"
race_data = pd.read_csv(url)
```

```python
race_data.isnull().sum()
```

```python
race_data.head(5)
```

```python
race_data['finish_pos'] = race_data['finish_pos'].map(lambda x: 1 if x >= 3 else 0)
```

```python
horse_data = race_data[['horse_roi', 'last_three_races', 'finish_pos']].copy()
horse_data
```

```python
y = horse_data[['finish_pos']].values 
```

```python
X = horse_data[['horse_roi', 'last_three_races']].values
```

```python
X_train, X, y_train, y = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
#splitting training into 70% and test and development into 15% 15%
X_dev, X_test, y_test, y_dev = train_test_split(X, y, test_size=0.50, random_state=1) 
```

```python
clf = SVC(kernel='poly',C=1.5,random_state=1)
clf.fit(X_train, y_train)
```

```python
score1 = clf.score(X_dev,y_dev)

print(score1)
```
